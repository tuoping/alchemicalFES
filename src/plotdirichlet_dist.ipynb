```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import dirichlet

def plot_dirichlet_2d(alpha):
    """
    Plot probability density of a 2-dimensional Dirichlet distribution.
    Args:
        alpha (list or np.array): alpha parameters, must be of length 2
    """
    if len(alpha) != 2:
        raise ValueError("This visualization only works for 2D Dirichlet (len(alpha) == 2).")

    # The 2D Dirichlet is equivalent to Beta(alpha[0], alpha[1])
    x = np.linspace(0, 1, 500)
    y = dirichlet.pdf(np.vstack([x, 1-x]), alpha)

    
    plt.plot(1-x, y, label=r"$\alpha$ =(%d, %d)"%(alpha[0], alpha[1]))
    plt.fill_between(1-x, y, alpha=0.3)


# Example usage:
plt.figure(figsize=(3.5, 3))
plot_dirichlet_2d([1, 1])
plot_dirichlet_2d([1, 5])
plot_dirichlet_2d([1, 10])
plt.xlabel("$x_{i1}$")
plt.ylabel("$p(x_{i\cdot})$")
plt.title(r"$g_{i}=(0,1)$")
plt.ylim((-1,10.5))
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import dirichlet

fig, ax1 = plt.subplots(figsize=(3.5, 3))

x = np.linspace(0, 1, 500)

# bottom axis (blue): label x_{i1}
y11 = dirichlet.pdf(np.vstack([x, 1-x]), [1, 1])
y1t = dirichlet.pdf(np.vstack([x, 1-x]), [1, 10])

ax1.plot(1 - x, y11, label=r"$\alpha=(1,1)$", c='tab:blue', alpha=0.6)
ax1.plot(1 - x, y1t, label=r"$\alpha=(1,t)$", c='tab:blue')
ax1.set_xlabel(r"$x_{i1}$", color='tab:blue')
ax1.set_ylabel(r"$p(x_{i\cdot})$", color='k')
ax1.tick_params(axis='x', colors='tab:blue')
ax1.set_ylim(0, 10.5)

# top axis (black): label x_{i0}
ax2 = ax1.twiny()
ax2.set_xlim(ax1.get_xlim())  # align x limits
ax2.plot(x,  y11, c='k', ls='--', alpha=0.5)  # repeat for legend/visual
ax2.plot(x,  y1t, c='k', ls='--')
ax2.set_xlabel(r"$x_{i0}$", color='k')

ax2.tick_params(axis='x', colors='k')

# one legend using handles from both axes
h1, l1 = ax1.get_legend_handles_labels()
h2, l2 = ax2.get_legend_handles_labels()
# ax1.legend(h1 + h2, l1 + l2, loc='upper center', ncols=2, frameon=False)

plt.tight_layout()
plt.show()
```
